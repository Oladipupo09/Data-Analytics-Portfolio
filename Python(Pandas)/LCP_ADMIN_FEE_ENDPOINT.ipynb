## Payment Endpoint

```python
# Get current year and month
today = datetime.today().date()
start_date = today.replace(day=1)
end_date = (today.replace(day=1) + timedelta(days=31)).replace(day=1) - timedelta(days=1)

# Convert dates to string format for JSON serialization
start_date = start_date.strftime('%Y-%m-%d')
end_date = end_date.strftime('%Y-%m-%d')

url = "https://data.upya.io/data/search/payments"

payload = json.dumps({
    "query": {
      
        "date": {
            "$gte": start_date, #'2025-03-01', #start_date,  # Ensure correct date format
            "$lte": end_date #'2025-03-31' #end_date
        }
    }
})

headers = {
    'Content-Type': 'application/json',
    'Authorization': 'Basic YXBpdXNlcjE6QWRtaW4xMjM0NSEq',
    'Cookie': 'connect.sid=s%3ArZm8O_FSzD56GnaoYEj5T0ceIdLP0b4C.IEDGFuQp5pxaWYNUYLmYFXXpQjfa%2BxQgfV21SGU8YFk'
}

payment = requests.request("POST", url, headers=headers, data=payload)

payment = payment.json()

payment
```

```python
# Step 2: Initialize an empty list to store the flattened data
flattened_data = []

# Step 3: Loop through each entry in the JSON file and flatten the data
for entry in payment:
    flat_entry = {
        'contract_type': entry.get('contract', {}).get('type', '') if entry.get('contract') else '', 
        'contractNumber': entry.get('contract', {}).get('contractNumber', '') if entry.get('contract') else '',
        'operator': entry.get('operator', ''),
        'type': entry.get('type', ''),
        'date': entry.get('date', ''),
        'amount': entry.get('amount', ''),
        'paymentReference': entry.get('paymentReference', ''),
        'ccy': entry.get('ccy', ''),
        'mobile': entry.get('mobile', ''),
        'status': entry.get('status', ''),
        'paymentCode': entry.get('paymentCode', ''),
        'transactionId': entry.get('transactionId', ''),
        
        # Ensure 'entity' exists before accessing subfields
        'entity_name': entry.get('entity', {}).get('name', '') if entry.get('entity') else '',
        'entity_number': entry.get('entity', {}).get('entityNumber', '') if entry.get('entity') else '',

        'days': entry.get('days', ''),
        'note': entry.get('note', ''),
    }
    
    # If there are nested 'collectBy' and 'recordBy' fields, extract them as well
    if entry.get('collectBy'):
        flat_entry['collectBy_firstName'] = entry['collectBy'].get('profile', {}).get('firstName', '')
        flat_entry['collectBy_lastName'] = entry['collectBy'].get('profile', {}).get('lastName', '')
        flat_entry['collectBy_agentNumber'] = entry['collectBy'].get('agentNumber', '')
    
    if entry.get('recordBy'):
        flat_entry['recordBy_firstName'] = entry['recordBy'].get('profile', {}).get('firstName', '')
        flat_entry['recordBy_lastName'] = entry['recordBy'].get('profile', {}).get('lastName', '')

    # Append the flattened entry to the list
    flattened_data.append(flat_entry)

# Step 4: Convert the flattened data into a DataFrame
df = pd.DataFrame(flattened_data)

pay = df[df['contract_type'] == 'PERPETUITY']

# Step 5: Write the DataFrame to an Excel file
pay.to_excel('payment_data.xlsx', index=False)

pay
```
